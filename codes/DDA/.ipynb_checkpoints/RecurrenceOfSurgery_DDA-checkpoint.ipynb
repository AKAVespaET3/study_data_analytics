### Quest 2
- 직업군이 상위 2위에 속하는 항목 : value_count, DataFrame, ilic
- 체중 > 70 , 신장 > 170, 흡연여부 없는분
- 해당 레코드 갯수 표현

```python
quest2_count = df_ROS_forpatient['직업'].value_counts().iloc[:2]
pd.DataFrame(quest2_count)
```

```python
quest2 = df_ROS_forpatient[['체중','신장','흡연여부']]
quest2_result = ((df_ROS_forpatient['직업'] == quest2_count.index[0]) | (df_ROS_forpatient['직업'] == quest2_count.index[1])) & (quest2['체중'] > 70) & (quest2['신장'] >170) & (quest2['흡연여부']==0)
quest2_result
```

```python
quest2_result_count = df_ROS_forpatient.loc[quest2_result]
quest2_result_count 
```

```python
len(df_ROS_forpatient.loc[quest2_result])
```
